## Resumen del proceso realizado

El proyecto comenzó con una inspección inicial del dataset original de usuarios de una plataforma de streaming.

Luego se realizó una etapa de calidad y limpieza, donde se trataron duplicados, inconsistencias categóricas, valores inválidos, fechas problemáticas y registros repetidos por `user_id`.

Posteriormente se desarrolló un análisis exploratorio de datos para estudiar la distribución de usuarios, el tiempo mensual de visualización y su relación con variables como plan de suscripción, país y género favorito.

Finalmente, se aplicó PCA sobre variables numéricas para analizar si era posible representar la información en menos dimensiones y visualizar la estructura general de los datos.

## Principales hallazgos

Durante la inspección inicial se detectaron problemas de calidad, como registros duplicados, `user_id` repetidos, valores faltantes, categorías inconsistentes, fechas inválidas y valores numéricos fuera de rango.

En la limpieza, se conservaron los datos originales en `data/raw` y se generó una versión procesada en `data/processed`. Las decisiones de transformación fueron documentadas en `logs/pipeline_log.csv`.

En el análisis exploratorio, se observó que el tiempo mensual de visualización fue una variable central para estudiar el comportamiento de los usuarios. La mayoría de los usuarios presenta consumos moderados, aunque existe una cola hacia la derecha con usuarios de consumo más alto.

El análisis por plan mostró que los usuarios Premium tienden a presentar mayores tiempos de visualización que los usuarios de planes Básico y Estándar.

El análisis por país mostró promedios de visualización bastante similares entre regiones, por lo que no se observaron diferencias fuertes en el consumo promedio según país.

El heatmap entre género favorito, plan y tiempo de visualización mostró que el plan Premium mantiene los valores promedio más altos en casi todos los géneros.

En PCA, las tres componentes explicaron porcentajes similares de varianza. Esto indica que la información numérica se encuentra distribuida de manera pareja entre `age`, `monthly_watch_time_mins` y `customer_support_tickets`. La proyección en PC1 y PC2 fue útil para visualizar, pero no separó claramente a los usuarios según el plan.

## Limitaciones del análisis

El dataset presentaba problemas de calidad que requirieron decisiones de limpieza, por lo que algunas variables fueron imputadas o marcadas como no válidas.

La variable `last_login_date` contenía fechas inválidas, faltantes y futuras. Estas fechas se marcaron como `NaT` en lugar de ser imputadas, ya que no correspondía inventar una fecha de último inicio de sesión.

El análisis realizado es exploratorio, por lo tanto no permite afirmar causalidad. Por ejemplo, aunque los usuarios Premium muestran mayor consumo promedio, no se puede afirmar que el plan cause directamente ese mayor consumo.

El PCA se aplicó únicamente sobre variables numéricas seleccionadas. No incluyó variables categóricas como país, género favorito o plan de suscripción.

## Conclusión final

El proyecto permitió analizar un dataset de usuarios de una plataforma de streaming siguiendo un proceso ordenado de minería de datos.

Primero se identificaron problemas de calidad, luego se aplicaron decisiones de limpieza justificadas y finalmente se exploraron patrones de comportamiento mediante visualizaciones y PCA.

El principal patrón observado fue que el tiempo mensual de visualización varía especialmente según el plan de suscripción. Los usuarios Premium presentan mayores niveles de consumo promedio, mientras que las diferencias por país resultan menos marcadas.

El análisis PCA permitió resumir la información numérica, aunque no mostró una separación clara de los usuarios según el plan de suscripción. Esto sugiere que las variables numéricas utilizadas no son suficientes por sí solas para diferenciar claramente los tipos de usuarios por plan.

En general, el análisis permitió transformar un dataset con problemas de calidad en una base más consistente y obtener hallazgos útiles sobre el comportamiento de visualización de los usuarios.
