```python
# Primeros valores de la columna last_login_date

df["last_login_date"].head(20)
```

La columna `last_login_date` representa la fecha del último inicio de sesión del usuario.

Aunque conceptualmente es una fecha, pandas la detecta como `object`, porque los valores están guardados como texto y presentan formatos mezclados o valores no válidos.

```python
# Valores más frecuentes en last_login_date

df["last_login_date"].value_counts(dropna=False).head(20)
```

Se revisan los valores más frecuentes de `last_login_date` para detectar posibles problemas de carga.

En esta etapa se busca identificar valores nulos, textos que no representan fechas, formatos distintos o fechas sospechosas.

```python
# Intentamos convertir last_login_date a formato fecha
# Los valores que no puedan convertirse quedarán como NaT

fechas_convertidas = pd.to_datetime(df["last_login_date"], errors="coerce")

fechas_convertidas.head()
```

Se realiza una conversión auxiliar de `last_login_date` a formato fecha.

El parámetro `errors="coerce"` permite que los valores que no puedan convertirse queden como `NaT`, es decir, valores de fecha no válidos.

Esta conversión no reemplaza la columna original. Solo se utiliza para diagnosticar qué valores presentan problemas.

```python
# Cantidad de fechas que no pudieron convertirse

fechas_invalidas = fechas_convertidas.isnull().sum()

print("Cantidad de fechas inválidas o faltantes:", fechas_invalidas)
```

```python
# Registros donde la fecha no pudo convertirse correctamente

df[fechas_convertidas.isnull()][["user_id", "last_login_date"]].head(20)
```

Los registros mostrados contienen valores que no pudieron interpretarse como fechas válidas.

Esto puede deberse a valores faltantes, textos como `not_available`, fechas imposibles como `0000-00-00`, o combinaciones de día, mes y año que no existen.

```python
# Tomamos como referencia la fecha actual
# Si una fecha de último login es futura, se considera sospechosa

fecha_actual = pd.Timestamp.today()

fechas_futuras = df[fechas_convertidas > fecha_actual]

fechas_futuras[["user_id", "last_login_date"]].head()
```

```python
# Cantidad de fechas futuras

len(fechas_futuras)
```

Se detectan fechas futuras en `last_login_date`.

Como la variable representa el último inicio de sesión, una fecha posterior a la fecha actual del análisis resulta sospechosa.  
Estas fechas no se corrigen en esta etapa, solo se documentan para tratarlas luego en el notebook de calidad y limpieza.

```python
# Edades menores o iguales a 0, o demasiado altas

edades_sospechosas = df[(df["age"] <= 0) | (df["age"] > 100)]

edades_sospechosas[["user_id", "age"]].head(20)
```

```python
# Cantidad de edades sospechosas

len(edades_sospechosas)
```

Se detectan edades sospechosas en la variable `age`, como valores menores o iguales a 0 o edades demasiado altas para el contexto del dataset.

Estos valores requieren atención porque pueden corresponder a errores de carga o registros inválidos.  
En esta etapa no se eliminan ni se modifican; solo se documentan para analizarlos en la etapa de limpieza.

```python
# Cantidad de tiempos de visualización sospechosos

len(tiempo_sospechoso)
```

Los valores negativos no tienen sentido para una variable que mide minutos de visualización.  
También aparecen valores extremadamente altos, que podrían ser errores de carga o casos atípicos que deben analizarse con mayor detalle.

En esta etapa no se corrigen estos valores; solo se registra la evidencia.
